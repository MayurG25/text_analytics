# text cleaning for summary extraction

```python
from nltk.tokenize import sent_tokenize,word_tokenize
from nltk.corpus import stopwords
from string import punctuation
```

```python
unwanted_words = ['e-mailshare','twittershare','scholarshare','e-mailpdf',
                 'facebookshare','pub','mednature.comgoogle',
                 '‘','’','—']
```

```python
# splitting the sentences
tokenised_sent = sent_tokenize(text)
```

```python
# splitting the words
tokenised_words = word_tokenize(text.lower())
```

```python
tokenised_words
```

```python
# removing stopwords,punctuations and additional words
_stopwords = set(stopwords.words('english')+ list(punctuation) + unwanted_words)
```

```python
filtered_words = [words for words in tokenised_words if  not words in _stopwords]
```

```python
# finding out most frequent words
from nltk.probability import FreqDist
```

```python
freq = FreqDist(filtered_words)
freq
```

```python
from heapq import nlargest
nlargest(15,freq,key=freq.get)
```

```python
# finding out significance score of the sentence
from collections import defaultdict
ranking = defaultdict(int)
```

```python
"""This step will give us significance score
   sentence significance = sum(frequency of words in the sentence) 
"""
for i,sent in enumerate(tokenised_sent): # enumerate converts a,b,c to tuples [(0,a),(1,b),(2,c)]
    for w in word_tokenize(sent.lower()):
        if w in freq:
            ranking[i] += freq[w]
```

```python
# significance score of each sentence which has most frequ
ranking
```

```python
# 56 sentences are available in the data
for i in enumerate(tokenised_sent):
    print(i)
```

```python
# finding out important sentences
sent_index = nlargest(4,ranking,key=ranking.get)
sent_index
```
